```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
export= pd.read_csv('International sales of oil and gas produced in Scotland (£billion).csv')
elec = pd.read_csv('Renewable Electricity.csv')
gdp = pd.read_csv('ScotGDP.csv')

export
```

```python
elec
```

```python
gdp
```

```python
gdp =  gdp[::-1].reset_index()

gdp = gdp.drop(columns=['index'])

gdp
```

```python
gdp['Onshore'] = 0.001 * gdp['Onshore']

gdp['Including Offshore'] = 0.001 * gdp['Including Offshore']

gdp
```

```python
elec = elec.drop(columns=['% Progress'])

elec
```

```python
export = export.drop(columns = ['Exports to the rest of the UK (£bn)', 'Exports to the rest of the World (£bn)'])

export
```

```python
df = pd.concat([gdp, export, elec], axis=1)

df
```

```python
#To drop the duplicated data
df = df.loc[:,~df.columns.duplicated()]

#To remove NaN rows
df = df.apply (pd.to_numeric, errors='coerce')
df = df.dropna()
df = df.reset_index(drop=True)


df
```

```python
# create figure and axis objects with subplots()
fig3,ax = plt.subplots(figsize=(15, 15), dpi=200)
fig3.suptitle("Scottish Exports of Oil and Gas Vs. Scottish GDP ")
ax.plot(df['Year'], df['Including Offshore'], color="black")
ax.set_ylim([90,200])
ax.set_ylabel("Scottish GDP: Billions of pounds")
plt.xticks(np.arange(2000, 2021, 1))

ax2 = ax.twinx()
ax2.set_ylabel("Total Scottish Exports: Billions of pounds")
ax2.plot(df['Year'], df['Total Exports (£bn)'], color="green")
ax2.set_ylim([5,30])

fig3.legend(labels  = [ "Exports of Oil and Gas", "Total GDP"], bbox_to_anchor=(0.3, 0.85, 0, 0))
```

```python
# create figure and axis objects with subplots()
fig2,ax = plt.subplots(figsize=(15, 15), dpi=200)
ax.plot(df['Year'], df['Renewable Electricity (GWh)'], color="#D4E824")
ax.set_ylim([20000,40000])
ax.set_ylabel("Renewable Electricity: GigaWatts per hour")
plt.xticks(np.arange(2000, 2021, 1))
ax.set_yticks(np.arange(0, 40000, 2000))
fig2.suptitle("Renewable Electricity in Scotland Vs. Scottish GDP ")

ax2 = ax.twinx()
ax2.plot(df['Year'], df['Including Offshore'], color="black")
ax2.set_yticks(np.arange(90, 190, 10))
ax2.set_ylabel("Scottish GDP: Billions of pounds")

fig2.legend(labels  = [ "Renewable Electricity", "Total GDP"], bbox_to_anchor=(0.3, 0.85, 0, 0))
```

```python
# create figure and axis objects with subplots()
fig1,ax = plt.subplots(figsize=(15, 5), dpi=200)
plt.xticks(np.arange(2000, 2021, 1))
ax.bar(df['Year']+0.35, df['Including Offshore'], 0.3, color="Blue")
ax.bar(df['Year'],df['Onshore'], 0.3, color="#34B04B")
ax.set_ylim([80,190])

ax.set_ylabel("GDP: Billions of Pounds")

fig1.suptitle("Scottish GDP: Onshore vs. Offshore sales")

ax.legend(labels  = ["Total GDP", "GDP With no Oil and Gas"])
```

```python
import plotly.express as px
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import dash_table
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def generate_table(dataframe, max_rows=26):
    return html.Table(
        # Header
        [html.Tr([html.Th(col) for col in dataframe.columns]) ] +
        # Body
        [html.Tr([
            html.Td(dataframe.iloc[i][col]) for col in dataframe.columns
        ]) for i in range(min(len(dataframe), max_rows))],
        style={
        'textAlign': 'center',
        'color': colors['text'],
        'margin-left': 'auto',
        'margin-right': 'auto',
        'font-family': 'Trebuchet MS'
        },
    )


# Build App
app = JupyterDash(__name__)


colors = {
    'background': '#111111',
    'text': '#FFFFFF'
}
app.layout = html.Div(style={'backgroundColor': colors['background']}, children=[
    html.H1(
        children='An Investigation of Scottish green energy against Scottish Exports and sales of Oil and Gas',
        style={
            'textAlign': 'center',
            'color': colors['text'],
            'font-family': 'Trebuchet MS'
        }
    ),
    html.H2(children='Created by Sam Martin', style={
        'textAlign': 'center',
        'color': colors['text'],
            'font-family': 'Trebuchet MS'
    }),
    html.Br(),

    html.Div(children=[
              html.H2(children='Scottish GPD over the years',
                     style={
                'textAlign': 'center',
                'color': colors['text'],
                'font-family': 'Trebuchet MS'
                  
            }),
              html.Div(children='The graph below has been produced to show that through the years from 2000-2020, the general trend is that the scottish GDP is increasing. A point of note to mention is how every single year there is a substantial increase when including the sales of oil and gas.', 
              style={
                'textAlign': 'center',
                'color': colors['text'],
                'font-family': 'Trebuchet MS'
                  
            })
    ]),
    dcc.Graph(
        id='Graph1',
        figure={
            'data': [
                {
                    'x': df['Year'], 
                    'y': df['Including Offshore'], 
                    'type': 'bar', 
                    'name': 'Total GDP', 
                    'marker' : { 'color': '#2E9BE0'},
                    'title_text': 'GDP in Billions of Pounds'
                },
                {'x': df['Year'], 'y': df['Onshore'], 'type': 'bar', 'name': 'GDP With no Oil and Gas', 'marker' : {'color': '#34B04B'}},
            ],
            'layout': {
                'plot_bgcolor': colors['background'],
                'paper_bgcolor': colors['background'],
                'font': {
                    'color': colors['text'],
            'font-family': 'Trebuchet MS'
                },
            
            },
            
        }
    ),
    html.Br(),
    html.Div(children=[
              html.H2(children="Renewable Energy and it's Effect on Scottish GPD!",
                     style={
                'textAlign': 'center',
                'color': colors['text'],
                'font-family': 'Trebuchet MS'
                  
            }),
              html.Div(children='The graph below has been produced to show that through the years from 2000-2020 as the amount of green energy being produced by scotland increases, the total Scottish GPD Increases as well.', style={
        'textAlign': 'center',
        'color': colors['text'],
            'font-family': 'Trebuchet MS'
            })
                           ]),
    dcc.RadioItems(
        id='radio1',
        value='Secondary'
    ),
    dcc.Graph(
        id="graph2"
        
    ),
    html.Br(),
    html.Div(children='The graph below has been produced to show that through the years from 2000-2020 both the Scottish GPD and Scotish Export Sales increase at the same time.', style={
        'textAlign': 'center',
        'color': colors['text'],
            'font-family': 'Trebuchet MS'
    }),
    dcc.RadioItems(
        id='radio2',
        value='Secondary'
    ),
    dcc.Graph(
        id="graph3"
        
    ),
    html.Br(),
    
    html.Div(children=[
              html.H2(children='Scottish Exports affecting the GPD?',
                     style={
                'textAlign': 'center',
                'color': colors['text'],
                'font-family': 'Trebuchet MS'
                  
            }),
    html.Div(children='Overall, this project shows that as the amount green energy produced in scotland increases, the scottish GPD and Exports have also increased. A key point to consider within this project that may show bias is how the fact that scottish GPD could also just increase as the country grows.', style={
        'textAlign': 'center',
        'color': colors['text'],
            'font-family': 'Trebuchet MS'
    })
    ]),
    html.Br(),
    html.Div(children=[
        html.H4(children='The dataset used to create the interactive graphs',
               style={
        'textAlign': 'center',
        'color': colors['text'],
            'font-family': 'Trebuchet MS'
        },),
        generate_table(df)
         
    ])
])



@app.callback(
    Output("graph2", "figure"), 
    [Input("radio1", "value")])
def display_(radio_value):

    # Create figure with secondary y-axis
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    
    # Add traces
    fig.add_trace(
        go.Scatter(x=df['Year'], y=df['Including Offshore'], name="Total GDP", line_color ='#22F63A'),
        secondary_y=False,
    )

    fig.add_trace(
        go.Scatter(x=df['Year'], y= df['Renewable Electricity (GWh)'], name="Renewable Electricity", line_color = '#D4E824'),
        secondary_y=radio_value == 'Secondary',
    )

    # Add figure title
    fig.update_layout(
        title_text="Total GDP of Scotland VS. Renewable Electricity produced by Scotland",
        plot_bgcolor = colors['background'],
                paper_bgcolor= colors['background'],
                font= {'color': colors['text']}
    )

    # Set x-axis title
    fig.update_xaxes(title_text="Years")

    # Set y-axes titles
    fig.update_yaxes(
        title_text="Total GDP of Scotland: Billions of Pounds", 
        secondary_y=False)
    fig.update_yaxes(
        title_text="Renewable Electricity (GWh)", 
        secondary_y=True)
    
 
          
    return fig

@app.callback(
    Output("graph3", "figure"), 
    [Input("radio2", "value")])
def display_(radio_value):

    # Create figure with secondary y-axis
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    
    # Add traces
    fig.add_trace(
        go.Scatter(x=df['Year'], y=df['Including Offshore'], name="Total GDP", line_color ='#22F63A'),
        secondary_y=False,
    )

    fig.add_trace(
        go.Scatter(x=df['Year'], y= df['Total Exports (£bn)'], name="Total Scottish Exports", line_color = '#2274F6'),
        secondary_y=radio_value == 'Secondary',
    )

    # Add figure title
    fig.update_layout(
        title_text="Total GDP of Scotland VS. Total Scottish Exports",
        plot_bgcolor = colors['background'],
                paper_bgcolor= colors['background'],
                font= {'color': colors['text']}
    )

    # Set x-axis title
    fig.update_xaxes(title_text="Years")

    # Set y-axes titles
    fig.update_yaxes(
        title_text="Total GDP of Scotland: Billions of Pounds", 
        secondary_y=False)
    fig.update_yaxes(
        title_text="Total Scottish Exports: Billions of Pounds", 
        secondary_y=True)
    
 
          
    return fig



# Run app and display result inline in the notebook
app.run_server(mode='inline')
```
